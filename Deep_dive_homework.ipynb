Checking if always the same combination is present or multiple combinations

```python
to_combine = [iso2, iso3, code3]
to_combine
```

```python
import itertools
```

```python
all_possible_combinations = [el for el in itertools.product(*to_combine)]
len(all_possible_combinations)
```

```python
all_possible_combinations[:5]
```

Combinations appearing in the data frame

```python
actual_combinations = []

for el in all_possible_combinations:
    _ = df[(df.iso2 == el[0]) & 
           (df.iso3 == el[1]) & 
           (df.code3 == el[2])].shape[0]
    
    if _ > 0:
        actual_combinations.append((el, _))
        
actual_combinations
```
